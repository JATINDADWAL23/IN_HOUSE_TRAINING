```python
import numpy as np
datascience_students_age=np.array([23,324,4,3,4,3,4,3])
java_student_age=np.array([23,34,34,54,54])
bins=[15,20,25,30,35,40,45]
plt.hist(datascience_students_age,bins,rwidth=0.8,histtype='bar',orientation='horizontal',color='r',label='datascience student')
plt.title('data science student')
plt.xlabel('student age')
plt.ylabel('no of students')
plt.legend(loc=8)
plt.show
```

```python
plt.hist(datascience_students_age,bins,rwidth=0.8,histtype='bar',orientation='vertical',color='magenta',label='datascience student')
plt.title('data science student')
plt.hist(java_student_age,rwidth=0.8,histtype='bar',orientation='vertical',color='aquamarine',label='Java Student age')
plt.title('data science student')
plt.xlabel('student age category')
plt.ylabel('no of student')
plt.legend(loc=3)
plt.show()
```
